Проверим зависимость времени за которые студент ответил на вопрос и корректностью ответа

```python
# Получаем выборку студентов, которые отвечали правильно
correct_answer_filt = df_train[(df_train['content_type_id'] == 0) & (df_train['answered_correctly'] == 1)]

# Получаем выборку студентов, которые отвечали правильно
ancorrect_answer_filt = df_train[(df_train['content_type_id'] == 0) & (df_train['answered_correctly'] == 0)]

# Выводим основные статистические свойства, получившихся выборок 
display(correct_answer_filt['prior_question_elapsed_time'].describe())
display(ancorrect_answer_filt['prior_question_elapsed_time'].describe())
```


Из полученных результатов видно, что в **среднем**, чтобы ответить правильно студенты необходимо потратить **больше времени**, чем студентам, которые не знали правильного ответа и отвечали неправильно (но также, нельзя исключать человеческий фактор и событие, при котором студент мог знать ответ, но по какой то причине оветил неправильно)

Также, можно заметить смещение **квантилей 25% и 75%** у полученных выборок.

Можно предположить, что смещение в большую сторону **25% квантиля** у студентов, которые отвечали на вопросы правильно, обусловлено более вдумчивым подходом к ответу на несложные вопросы (Возьмем градацию сложности вопроса по времение ответа на него, чем сложнее вопрос тем больше времени на него тратит студент).
Тогда как студенты, которые ответили неправильно, скорее всего под влиянием стресса и легкой паники из за незнания ответа, старались как можно быстрее "избавиться" от этого чувства, по средствам ответа на вопрос, пускай даже и неправильного, так сказать "На удачу"

**Смещение 75% квартиля**, в меньшую сторону у студентов, отвечавших правильно, связываю с тем что они более быстро справлялись с разбором и понимаем сложных вопросов 
благодаря этому отвечали немного быстрее, чем вторая группа студентов.

Теперь проверим 

```python
content_type_id_true = df_train[(df_train['content_type_id'] == 1)]
students_read_the_lecture = df_train[(df_train['user_id'].isin(content_type_id_true['user_id'])) & (df_train['content_type_id'] == 0)]
students_didnt_read_the_lecture = df_train[(~df_train['user_id'].isin(content_type_id_true['user_id']))]
display('Количество студентов в 1 выборке', students_read_the_lecture['user_id'].nunique())
display(students_read_the_lecture.describe())
display('Количество студентов в 2 выборке', students_didnt_read_the_lecture['user_id'].nunique())
display(students_didnt_read_the_lecture.describe())
```

На основе получившихся данных можно получить предварительный вывод, что прочтение студентами лекций положительно сказывается на их ответах на вопросы.

**66% правильно** отвеченных вопросов и **859485** в принципе отвеченных вопросов было сделано **1475 студентами**.
Тогда как, **57.5% правильно** отвеченных вопросов и всего лишь **120608** в принципе отвеченных вопросов было сделано **2349 студентами**.

То есть в **1,6** раза меньшее количество студентов, ответило **в 7** раз больше вопросов с процентом правильных оветов **66%**.
